```python
d1 = Normal(μ=0, σ=1)
# d1 = Exponential(λ=5)
bs = np.linspace(1, 3, 101)
bsq85 = np.array([d1.BSQ(lambda x: x ** b, lambda x: x ** 1 / b, 0.85) for b in bs])
bsq90 = np.array([d1.BSQ(lambda x: x ** b, lambda x: x ** 1 / b, 0.90) for b in bs])
bsq95 = np.array([d1.BSQ(lambda x: x ** b, lambda x: x ** 1 / b, 0.95) for b in bs])
```

```python
plt.plot(bs, bsq85, label='85')
plt.plot(bs, bsq90, label='90')
plt.plot(bs, bsq95, label='85')
plt.legend()
```
